```python
import yaml
import os

dataset_info = {
    'train': '/content/ultralytics/Helmet_detect_dataset/images/train',
    'val': '/content/ultralytics/Helmet_detect_dataset/images/val',
    'path': '/content/ultralytics/Helmet_detect_dataset',
    'nc': 4,
    'names': ['helmet', 'vest', 'head', 'person']
}

yaml_filepath = '/content/ultralytics/Helmet_detect_dataset/data.yaml'

os.makedirs(os.path.dirname(yaml_filepath), exist_ok=True)

with open(yaml_filepath, 'w') as f:
    yaml.dump(dataset_info, f, default_flow_style=None, sort_keys=False)

with open(yaml_filepath, 'r') as f:
    loaded_yaml = yaml.safe_load(f)
    print(loaded_yaml)
```
